```
# bring in our packages for SLR
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model, metrics
from scipy import stats

# load data set
rent_df = pd.read_csv('https://raw.githubusercontent.com/mlhwang/ma322/master/rent.csv')
```

### Lab Question 1:

Target Feature: Rent

It is a numerical feature.

Regression is used to predict numerical features.

### Lab Question 2:


```
print(plt.scatter(rent_df['size'], rent_df['rent']))
```

```
# change shape/color
print(plt.scatter(rent_df['size'], rent_df['rent'], marker='+', color='turquoise'))
```

```
# bring it all together
plt.scatter(rent_df['size'], rent_df['rent'], color='maroon', marker='.')
plt.xlabel("Size")
plt.ylabel("Rent")
plt.title("Scatter Plot of Rent vs Size")
```

### Lab Question 3:


```
print(stats.pearsonr(rent_df['size'], rent_df['rent']))
```

The Pearson’s correlation coefficient, r , is `0.90388`.

From this we know that size and rent have a strong positive relationship, which is consistent with the results of the scatter plot.

### Lab Question 4:


```
model = linear_model.LinearRegression().fit(rent_df[['size']], rent_df['rent'])

print("The coefficient is:", model.coef_)
print("The intercept is:", model.intercept_)
```

Ŷ = 201.38 + 1.06*X

In context:

rent_hat = 201.38 + 1.06*size

### Lab Question 5:


```
# storing the predictions
rent_predictions = model.predict(rent_df[['size']])

# calculating the R^2
metrics.r2_score(rent_df['rent'], rent_predictions)
```

The R^2 value is .817.

This indicates that 81.7% of the variation in rent is explained by this
model.

### Lab Question 6:


```
# making predictions using our model
print("Our model predicts that 950 square feet would cost:", model.predict([[950]]))
print("Our model predicts that 1500 square feet would cost:", model.predict([[1500]]))
```

###Lab Question 7:

```
# graphing the linear regression line

# re-plot the original data (as points)
plt.plot(rent_df['size'],rent_df['rent'], 'o', marker='*', color='lime' )

# plot the data of size as X, and the predictions as Y (as a line)
plt.plot(rent_df['size'], rent_predictions, '-', linewidth=3, color='green')
```
