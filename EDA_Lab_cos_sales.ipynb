### Question 2:

```
print(pd.crosstab(index=cos['salesperson'], columns='count'))  # 2a.
```

```
# 2b
rel_freq_tab = pd.crosstab(index=cos['dept'], columns='count')
print(rel_freq_tab/rel_freq_tab.sum())
```

```
# 2c
# so first we need a frequency table to count
freq_month = pd.crosstab(index=cos['month'], columns="count")

# and then we'll put that frequency calculation into a bar plot
print(freq_month.plot.bar())
```

```
# 2d
freq_dept = pd.crosstab(index=cos['dept'], columns='count')
print(freq_dept.plot.pie(y="count"))
```

```
# 3b

import statistics

print("The mean of order amounts is:", cos['order_amount'].mean())
print("The mean of units sold is:", cos['units_sold'].mean())
print("The median of order amounts is:", cos['order_amount'].median())
print("The median of units sold is:", cos['units_sold'].median())
print("The standard deviation of order amounts is:", statistics.stdev(cos['order_amount']))
print("The standard deviation of units sold is:", statistics.stdev(cos['units_sold']))
```
